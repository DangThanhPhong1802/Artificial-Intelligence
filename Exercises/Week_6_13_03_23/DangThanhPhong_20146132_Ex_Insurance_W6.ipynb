Dùng file 'insurance.csv'

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model #thư viện dùng cho machine learning - mô hình tuyến tính

url = 'insurance.csv'
data = pd.read_csv(url)
data.hist()  #vẽ phân bố dữ liệu Histogram
plt.show
```

```python
sns.pairplot(data)
```

```python
correlation = data.corr()
print(correlation)

sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind='scatter', x='age', y='charges', figsize=(9,9), color='black')
```

```python
#Hồi quy hàm bậc 1
model = linear_model.LinearRegression()
X=pd.DataFrame(data['age'])
y=data['charges']
model.fit(X, y)
print('Intercept của mô hình : ', model.intercept_)
print('Coefient của mô hình : ', model.coef_)
print('Độ chính xác của mô hình : ', model.score(X,y))

range = np.arange(data['age'].min(), data['age'].max(),0.1)
data.plot(kind='scatter', x='age', y='charges', figsize=(9,9), color='black', xlim=0.7)
predictor = pd.DataFrame(range)
y_pred = model.predict(X=predictor)
plt.plot(range, y_pred, color='blue')
```

```python
#Hồi quy hàm bậc 5
model = linear_model.LinearRegression()
X=pd.DataFrame([data['age'], data['age']**2, data['age']**3, data['age']**4, data['age']**5]).T
y=data['charges']
model.fit(X, y)
print('Intercept của mô hình : ', model.intercept_)
print('Coefient của mô hình : ', model.coef_)
print('Độ chính xác của mô hình : ', model.score(X,y))

range = np.arange(data['age'].min(), data['age'].max(),0.1)
data.plot(kind='scatter', x='age', y='charges', figsize=(9,9), color='black', xlim=0.7)
predictor = pd.DataFrame([range, range**2, range**3, range**4, range**5]).T
y_pred = model.predict(X=predictor)
plt.plot(range, y_pred, color='blue')
```
