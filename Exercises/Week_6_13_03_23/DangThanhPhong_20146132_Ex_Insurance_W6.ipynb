```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model #thÆ° viá»n dÃ¹ng cho machine learning - mÃ´ hÃ¬nh tuyáº¿n tÃ­nh

url = 'insurance.csv'
data = pd.read_csv(url)
data.hist()  #váº½ phÃ¢n bá» dá»¯ liá»u Histogram
plt.show
```

```python
sns.pairplot(data)
```

```python
correlation = data.corr()
print(correlation)

sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind='scatter', x='age', y='charges', figsize=(9,9), color='black')
```

```python
#Há»i quy hÃ m báº­c 1
model = linear_model.LinearRegression()
X=pd.DataFrame(data['age'])
y=data['charges']
model.fit(X, y)
print('Intercept cá»§a mÃ´ hÃ¬nh : ', model.intercept_)
print('Coefient cá»§a mÃ´ hÃ¬nh : ', model.coef_)
print('Äá» chÃ­nh xÃ¡c cá»§a mÃ´ hÃ¬nh : ', model.score(X,y))

range = np.arange(data['age'].min(), data['age'].max(),0.1)
data.plot(kind='scatter', x='age', y='charges', figsize=(9,9), color='black', xlim=0.7)
predictor = pd.DataFrame(range)
y_pred = model.predict(X=predictor)
plt.plot(range, y_pred, color='blue')
```
