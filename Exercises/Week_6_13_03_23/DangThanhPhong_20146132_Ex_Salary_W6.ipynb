Dùng file 'Salary_Data.csv'

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model #thư viện dùng cho machine learning - mô hình tuyến tính

url = 'Salary_Data.csv'
data = pd.read_csv(url)
data.hist()  #vẽ phân bố dữ liệu Histogram
plt.show
```

```python
sns.pairplot(data)
```

```python
correlation = data.corr()
print(correlation)
```

```python
sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind='scatter', x='Salary', y='YearsExperience', figsize=(9,9), color='black')
```

```python
#Hồi quy hàm bậc 1
model = linear_model.LinearRegression()
X=pd.DataFrame(data['Salary'])
y=data['YearsExperience']
model.fit(X, y)
print('Intercept của mô hình : ', model.intercept_)
print('Coefient của mô hình : ', model.coef_)
print('Độ chính xác của mô hình : ', model.score(X,y))

range = np.arange(data['Salary'].min(), data['Salary'].max(),0.1)
data.plot(kind='scatter', x='Salary', y='YearsExperience', figsize=(9,9), color='black', xlim=0.7)
predictor = pd.DataFrame(range)
y_pred = model.predict(X=predictor)
plt.plot(range, y_pred, color='blue')
```

```python
#Hồi quy bậc 2
model = linear_model.LinearRegression()
X=pd.DataFrame([data['Salary'], data['Salary']**2]).T
y=data['YearsExperience']
model.fit(X, y)
print('Intercept của mô hình : ', model.intercept_)
print('Coefient của mô hình : ', model.coef_)
print('Độ chính xác của mô hình : ', model.score(X,y))

range = np.arange(data['Salary'].min(), data['Salary'].max(),0.1)
data.plot(kind='scatter', x='Salary', y='YearsExperience', figsize=(9,9), color='black', xlim=0.7)
predictor = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X=predictor)
plt.plot(range, y_pred, color='blue')
```

```python
#Hồi quy bậc 3
model = linear_model.LinearRegression()
X=pd.DataFrame([data['Salary'], data['Salary']**2, data['Salary']**3]).T
y=data['YearsExperience']
model.fit(X, y)
print('Intercept của mô hình : ', model.intercept_)
print('Coefient của mô hình : ', model.coef_)
print('Độ chính xác của mô hình : ', model.score(X,y))

range = np.arange(data['Salary'].min(), data['Salary'].max(), 0.1)
data.plot(kind='scatter', x='Salary', y='YearsExperience', figsize=(9,9), color='black', xlim=0.7)
predictor = pd.DataFrame([range, range**2, range**3]).T
y_pred = model.predict(X=predictor)
plt.plot(range, y_pred, color='blue')
```
