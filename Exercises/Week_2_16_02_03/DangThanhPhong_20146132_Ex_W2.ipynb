```python
#Khai báo thư viện
import pandas as pd
import numpy as np
```

```python
#Khai báo đường dẫn
URL = 'http://winterolympicsmedals.com/medals.csv'
#Đọc dữ liệu
my_df = pd.read_csv(URL)
```

```python
#Hiển thị số lượng dòng và cột của file csv
print('(Dòng, cột) :', my_df.shape)
#Hiển thị tổng số lượng phần tử của file csv
print('SL phần tử :', my_df.size)
```

```python
#Hiển thị 1 số thông tin của file csv :
  #Index của bảng
  #Tên các cột, số lượng các phần tử Null trong cột và kiểu dữ liệu của chúng
  #Số lượng các kiểu dữ liệu
  #Dung lượng của bảng
my_df.info()
```

```python
#Thống kê đơn giản như count, mean, std, min, max
#Ngoài ra còn có : percentiles = [0.25, 0.5, 0.75]
#Phương thức này chỉ áp dụng cho các cột ở dạng numerical
my_df.describe()
```

```python
#Có thể thay đổi percentiles bằng cách truyền tham số này vào trong .describe()
#Lưu ý : pandas mặc định tính thêm percentile tại 0.5 dù không truyền vào
my_df.describe(percentiles=[0.1,0.85])
```
