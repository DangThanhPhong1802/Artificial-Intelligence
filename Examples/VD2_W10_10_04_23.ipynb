```python
from os import listdir
from numpy import asarray, save
#from keras.preprocessing.image import load_img, img_to_array
from keras.utils import load_img, img_to_array
```

```python
folder = '/content/drive/MyDrive/image_train/'
photos, labels = list(), list()
for file in listdir(folder) :
  output = 0.0
  if file.startswith('Phong') :
    output = 1.0
  if file.startswith('Phu') :
    output = 2.0
  photo = load_img(folder+file, target_size=(200,200))
  photo = img_to_array(photo)
  photos.append(photo)
  labels.append(output)

photos = asarray(photos)
labels = asarray(labels)
print(photos.shape, labels.shape)
save('/content/drive/MyDrive/Saved_Train/Phong_Phu_photos.npy', photos)
save('/content/drive/MyDrive/Saved_Train/Phong_Phu_labels.npy', labels)
```
